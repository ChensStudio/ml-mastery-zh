```python
from pandas_datareader import wb
import pandas as pd
pd.options.display.width = 0

names = [
    "NE.EXP.GNFS.CD", # Exports of goods and services (current US$)
    "NE.IMP.GNFS.CD", # Imports of goods and services (current US$)
    "NV.AGR.TOTL.CD", # Agriculture, forestry, and fishing, value added (current US$)
    "NY.GDP.MKTP.CD", # GDP (current US$)
    "NE.RSB.GNFS.CD", # External balance on goods and services (current US$)
]

df = wb.download(country="all", indicator=names, start=2010, end=2010).reset_index()
countries = wb.get_countries()
non_aggregates = countries[countries["region"] != "Aggregates"].name
df_nonagg = df[df["country"].isin(non_aggregates)].dropna()
print(df_nonagg)
```

```python
vectors = {}
for rowid, row in df_nonagg.iterrows():
    vectors[row["country"]] = row[names].values

print(vectors)
```

```python
import numpy as np

euclid = {}
cosine = {}
target = "Australia"

for country in vectors:
    vecA = vectors[target]
    vecB = vectors[country]
    euclid[country] = np.linalg.norm(vecA - vecB)
    cosA = vecA / np.linalg.norm(vecA)
    cosB = vecB / np.linalg.norm(vecB)
    cosAB = np.dot(cosA, cosB)
    cosine[country] = 1- cosAB
    # consine[country] = 1 - (vecA @ vecB) / (np.linalg.norm(vecA) * np.linalg.norm(vecB))
# # print("Euclidean distances from Australia:")
# # for country in sorted(euclid, key=euclid.get):
# #     print(f"{country}: {euclid[country]:.2f}")
# print("Cosine distances from Australia:")
# for country in sorted(cosine, key=cosine.get):
#     print(f"{country}: {cosine[country]:.2f}")
# The code above uses the World Bank API to fetch economic data for all countries
# and calculates the Euclidean and cosine distances between Australia and other countries
# based on selected economic indicators. The results are printed in sorted order.
```

```python
import pandas as pd

df_distances = pd.DataFrame({
    "euclidean": euclid,
    "cosine": cosine
})

print(df_distances.sort_values("euclidean").head())
```
